## Plot semplice

```python
import matplotlib.pyplot as plt
import numpy as np

# Funzione da plottare
x = np.arange(0.,8.*np.pi,0.01) # Asse x, np array
y = np.sin(x)

# Plot
fig, ax = plt.subplots(figsize=(8,6)) # Inizializzazione oggetti
ax.plot(x,y) # Plot
plt.show()
```

```python
from scipy.stats import cauchy
x_lor = np.arange(-5.,5.,0.01)
y_lor = np.zeros(len(x_lor))
weights  = [1.2,0.8]
widths   = [0.2,1]
energies = [0,2]
for i in range(len(weights)): 
    y_lor+=weights[i]*cauchy.pdf(x_lor,energies[i],widths[i])

fig, ax = plt.subplots(figsize=(8,6)) # Inizializzazione oggetti

# Aggiungiamo limiti del plot, linee verticali e coloriamo l'area del sottografico
ax.set_xlim(-5,5)
ax.set_ylim(0.,np.max(y_lor)*1.1)
ax.axvline(0,ls='--',c='black',lw=1.2)
ax.axvline(2,ls='--',c='black',lw=1.2)

ax.fill_between(x_lor,y_lor,alpha=0.4,color='blue')
ax.plot(x_lor,y_lor,'-',c='blue')

plt.show()
```

### Colormap
Lista delle [colormap](https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html) disponibili.

```python
# Funzione da plottare
x   = np.arange(0.,8.*np.pi,0.01)    #Generazione array con step
phases = np.linspace(0.,2.*np.pi,10) # Generazione array con numero di elementi

y = np.sin(x)

functions = [] #Ogni funzione ha una differenza di fase di 2pi/10 dalla precedente
for phase in phases: functions.append(np.sin(x+phase)) 
    
# Plot
fig, ax = plt.subplots(figsize=(8,6)) # Inizializzazione oggetti

xticks = [ i*np.pi for i in range(9)] # Inline for loop
xlabels = [ str(i)+'π' for i in range(9) ]

yticks = [-1.,-0.5,0.,0.5,1.]
ylabels = yticks

ax.set_xticks(xticks)
ax.set_xticklabels(xlabels,fontsize=18)
ax.set_yticks( yticks )
ax.set_yticklabels(ylabels,fontsize=18)

ax.set_xlabel('angolo (rad)',fontsize=18)
ax.set_ylabel('sin (a.u.)',fontsize=18)
ax.set_title('Plot semplice',fontsize=20)

# linewidth->lw, linestyle->ls, color->c, etc
for function in functions: ax.plot(x,function,'-',lw=1)

plt.show()
```

```python
# Utilizziamo una color map
colormap = plt.cm.rainbow #plt.cm.jet, plt.cm.[VEDI LINK SOPRA]
# Selezioniamo un colore per ogni funzione
# (le colormap sono normalizzate da 0 a 1)
colors = [colormap(i) for i in np.linspace(0, 1, 10)] 

# Plot
fig, ax = plt.subplots(figsize=(8,6)) # Inizializzazione oggetti

xticks = [ i*np.pi for i in range(9)] # Inline for loop
xlabels = [ str(i)+'π' for i in range(9) ]

yticks = [-1.,-0.5,0.,0.5,1.]
ylabels = yticks

ax.set_xticks(xticks)
ax.set_xticklabels(xlabels,fontsize=18)
ax.set_yticks( yticks )
ax.set_yticklabels(ylabels,fontsize=18)

ax.set_xlabel('angolo (rad)',fontsize=18)
ax.set_ylabel('sin (a.u.)',fontsize=18)
ax.set_title('Plot semplice',fontsize=20)

for i,function in enumerate(functions): ax.plot(x,function,'-',lw=1.5,c=colors[i])

# Aggiunta di COLORBAR: super difficile per questo tipo di plot! 
# In questi casi bisogna cercare lo 'snippet' di codice giusto online!
from matplotlib import colors
norm = colors.Normalize(vmin=0,vmax=2.*np.pi)
sm = plt.cm.ScalarMappable(cmap=colormap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ticks=np.linspace(0,2.*np.pi,len(functions)), 
             boundaries=np.arange(-0.05,2.*np.pi,0.1))

plt.show()
```

### Plot di funzione a più variabili
$z=f(x,y)$

```python
from scipy.stats import norm

# Generiamo i dati sommando due distribuzioni normaly su una mesh (x,y)

N = 40
x = np.linspace(0,1,N)
y = np.linspace(0,1,N)
xx, yy = np.meshgrid(x,y) # Creiamo una mesh 2D su cui valutare la funzione

mean  = 0.5
sigma = 0.3
# Sommiamo due distribuzioni normali
z = norm.pdf(xx,loc=mean,scale=sigma)+norm.pdf(yy,loc=mean,scale=sigma)
#print(z)

# Plot
fig, ax = plt.subplots(figsize=(8,8)) # Inizializzazione oggetti

# Color map
colormap = plt.cm.viridis

ax.set_title('gaussiana 2D',fontsize=18)

ax.set_xlim(0,1)
ax.set_ylim(0,1)

ticks = [0.,0.2,0.4,0.6,0.8,1.]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(ticks,fontsize=15)
ax.set_yticklabels(ticks,fontsize=15)

ax.set_xlabel('x',fontsize=18)
ax.set_ylabel('y',fontsize=18)

# I valori di z appaiono come una scala di colori
# Possiamo anche mappare z nella dimensione dei punti

#ax.scatter(xx, yy,marker='.',s=300,c='red',edgecolor='black') #Mesh 2d regolare
#ax.scatter(xx, yy,marker='s',s=20,c=z,cmap=colormap)
ax.scatter(xx, yy,marker='s',s=30*z,c=z,cmap=colormap)

plt.show()
```

## Istogramma
$p(x) = \frac{1}{\sqrt{ 2 \pi \sigma^2 }}
e^{ - \frac{ (x - \mu)^2 } {2 \sigma^2} }$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generiamo i dati 
N = 500
mean = 0    # media
sigma = 15  # dev st
x = np.random.normal(mean,sigma,N) #Genera dati secondo una distribuzione gaussiana

num_bins = 50

fig, ax = plt.subplots(figsize=(8,6))

# Istogramma
ax.set_xlim(-50,50)
#ax.hist(x, num_bins)
counts, bins, _ = ax.hist(x, num_bins, color='blue', density=True) #density=True: normalizzato a 1

# Facciamo un fit gaussiano
y =  np.exp(-0.5 * ( (bins - mean)/sigma )**2) / (sigma*np.sqrt(2. * np.pi)) 

# Controlliamo con una funzione
x_smooth = np.linspace(-100,100,1000)
y_true = norm.pdf(x_smooth,loc=0,scale=15)

ax.plot(bins, y, '-',lw=5, color='red',label='fit',zorder=1)
ax.plot(x_smooth, y_true, '--',lw=5, color='black',label='true function')

ax.set_xlabel('Bins')
ax.set_ylabel('Conteggi')
ax.set_title(r'Istogramma $\mu=100$, $\sigma=15$') #Lettere greche

plt.legend()
plt.show()
```
