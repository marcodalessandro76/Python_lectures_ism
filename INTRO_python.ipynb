```python
# Costruzione di un dataset medio e analisi con plot
from glob import glob
import numpy as np
import matplotlib.pyplot as plt

filenames = sorted(  glob('data/inflammation*.csv')   )

N_pazienti, N_giorni = [60, 40]

# Dataset medio
average_set = np.zeros(  (N_pazienti,N_giorni)  )

for filename in filenames:
    data = np.loadtxt(fname=filename, delimiter=',')
    average_set = average_set + data

average_set = average_set/len(filenames)

# Plot dataset medio
fig = plt.figure(figsize=(8,6))
plt.title('Dataset medio')
plt.imshow(average_set)
plt.show()

# Plot singolo dataset (12esimo)
fig = plt.figure(figsize=(8,6))
plt.title('Singolo dataset')
plt.imshow(data)
plt.show()

# Media pazienti del dataset medio
fig = plt.figure(figsize=(4,3))
plt.title('Media giornaliera')
plt.plot( np.mean(average_set,axis=0), c='blue'  )
# Media pazienti del singolo dataset
plt.plot( np.mean(data,axis=0), c='red' )
plt.show()
```

```python
# Funzione
def fahr_to_celsius(temperatura):
    return ((temperatura-32)*5./9.)

# Corpo principale con chiamata alla funzione definita sopra
print('Bolle a:', fahr_to_celsius(212),'C')
```

```python
# Funzione
def fahr_to_celsius(temperatura,grado='F'):
    """
    Conversione da gradi Fahrenheit a gradi Celsius e viceversa
    
    - temperatura: temperatura
    - conversione F->C: grado='F'
    - conversione C->F: grado='C'
    """
    if grado=='F':   new_temp = ((temperatura-32)*5./9.)
    elif grado=='C': new_temp = temperatura*9./5.+32
    else: return 'ERROR'
   
    return new_temp

# Corpo principale
print('Bolle a:', fahr_to_celsius(212),'C')
print('Bolle a:', fahr_to_celsius(100,grado='C'), 'F')
print('Bolle a:', fahr_to_celsius(100,grado='CF'), 'F')
```

```python
# Funzioni
def fahr_to_kelvin(temp):
    """
    Conversione da gradi Fahrenheit a kelvin
    """
    temp_celsius = fahr_to_celsius(temp)
    return temp_celsius+273.15

def fahr_to_celsius(temperatura,grado='F'):
    """
    Conversione da gradi Fahrenheit a gradi Celsius e viceversa
    
    - temperatura: temperatura
    - conversione F->C: grado='F'
    - conversione C->F: grado='C'
    """
    if grado=='F':   new_temp = ((temperatura-32)*5./9.)
    elif grado=='C': new_temp = temperatura*9./5.+32
    else: return 'ERROR'
   
    return new_temp

# Corpo principale
temp_kelvin = fahr_to_kelvin(212)
print(temp_kelvin)
```

```python
# Funzioni che ritornano più di un argomento
def fahr_to_kelvin(temp):
    """
    Conversione da gradi Fahrenheit a kelvin
    """
    temp_celsius = fahr_to_celsius(temp)
    return temp_celsius+273.15, temp_celsius # Qui la funzione ritorna sia la temp. in kelvin, sia in gradi Celsius

def fahr_to_celsius(temperatura,grado='F'):
    """
    Conversione da gradi Fahrenheit a gradi Celsius e viceversa
    
    - temperatura: temperatura
    - conversione F->C: grado='F'
    - conversione C->F: grado='C'
    """
    if grado=='F':   new_temp = ((temperatura-32)*5./9.)
    elif grado=='C': new_temp = temperatura*9./5.+32
    else: return 'ERROR'
   
    return new_temp

# Corpo principale
temp_kelvin, temp_celsius = fahr_to_kelvin(212) # Leggiamo entrambi gli output
print(temp_kelvin, temp_celsius)

temperatures = fahr_to_kelvin(212) # Anche così leggiamo entrambi gli output
print(temperatures)

temp_kelvin2, _ = fahr_to_kelvin(212)    # Leggiamo solo il primo output
print(temp_kelvin2)

temp_kelvin3    = fahr_to_kelvin(212)[0] # Anche così leggiamo solo il primo output
print(temp_kelvin3)
```

### Funzione con diverse operazioni

```python

def fai_plot(filename):
    """
    Plotta media, max e min di un dataset
    """
    
    data = np.loadtxt(fname=filename, delimiter=',')
    
    # Inizializzazione figura
    fig = plt.figure( figsize=(10, 3) )

    subplt1 = fig.add_subplot(1,3,1)
    subplt2 = fig.add_subplot(1,3,2)
    subplt3 = fig.add_subplot(1,3,3)

    subplt1.set_ylabel('Media')
    subplt2.set_ylabel('Max')
    subplt3.set_ylabel('Min')

    subplt1.plot( np.mean(data, axis=0) )
    subplt2.plot( np.max(data, axis=0)  )
    subplt3.plot( np.min(data, axis=0), c='red' )

    fig.tight_layout()  # Avvicina tutti i subplot
    plt.show()

# Corpo principale
filenames = sorted( glob('data/inflammation*.csv') )
for filename in filenames[:3]:
    print(filename)
    fai_plot(filename)
```

```python
# Formulazione più concisa di fai_plot
import numpy as np
import matplotlib.pyplot as plt
from glob import glob

def fai_plot(filename):
    """
    Plotta media, max e min di un dataset
    
    NOTA: questa funzione dovrebbe contenere una parte di 'error handling'
          per assicurarci che venga utilizzata nel modo che intendiamo...
    """  
    data = np.loadtxt(fname=filename, delimiter=',') # Lettura database
    
    # Inizializzazione figura
    fig = plt.figure( figsize=(10, 3) )
    
    num_plots = 4 # Numero di plot
    axes = []     # Lista di subplot
    for i in range(num_plots): axes.append( fig.add_subplot(1, num_plots, i+1) ) # Riempiamo la lista
        
    labels = ['Media', 'Max', 'Min', 'Dev. St.'] # Assi y dei vari subplot
    data_to_plot = [ np.mean(data, axis=0), np.max(data, axis=0), \
                     np.min(data, axis=0), np.std(data, axis=0) ] # Contenuto dei vari subplot
    
    for i,ax in enumerate(axes): # Facciamo il plot
        ax.set_ylabel( labels[i] )
        ax.plot( data_to_plot[i] )
    
    fig.tight_layout()
    plt.show() # Mostriamo il plot a schermo

# Corpo principale con chiamata alla funzione sopra definita

if __name__ == "__main__": # Istruzione opzionale (da mettere se lo script viene importato in python da altre parti)
    
    filenames = sorted( glob('data/inflammation*.csv') )
    for filename in filenames[:3]: # Ciclo soltanto sui primi 3 filename
        print(filename)
        fai_plot(filename)    
```
